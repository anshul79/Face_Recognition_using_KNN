```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
import os
```

```python
files = os.listdir('pic_data')
# print files
faces_paths = []
labels=[]

for f in files:
    ims = os.listdir('pic_data' + '/' + f)
    # print f, ims
    for px in ims:
        faces_paths.append('pic_data' + '/' + f + '/' + px)
        labels.append(f)

print faces_paths
print labels
```

```python
imgs = []

for fx in faces_paths:
    im = cv2.imread(fx)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    imgs.append(im)

print len(imgs)
```

```python
all_faces=[]
for i in range(0,5):
    plt.figure(i)
    im = imgs[i][20:300,100:350,:]
    print im.shape
    plt.imshow(im)

    img = cv2.resize(im, (100, 100))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    print img.shape
    plt.imshow(img, cmap='gray')
    plt.show()

    all_faces.append(img)
```

```python
for i in range(5,10):
    plt.figure(i)
    im = imgs[i][20:300,100:320,:]
    print im.shape
    plt.imshow(im)

    img = cv2.resize(im, (100, 100))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    print img.shape
    plt.imshow(img, cmap='gray')
    plt.show()

    all_faces.append(img)
```

```python
for i in range(10,15):
    plt.figure(i)
    im = imgs[i][40:350,200:400,:]
    print im.shape
    plt.imshow(im)

    img = cv2.resize(im, (100, 100))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    print img.shape
    plt.imshow(img, cmap='gray')
    plt.show()

    all_faces.append(img)
```

```python
print len(all_faces)
all_faces=np.asarray(all_faces)
print all_faces.shape
```

```python
lab = np.asarray(labels[:15])
print lab.shape

np.save('pic_data/my_face_labels', lab)
np.save('pic_data/my_face_data', all_faces)
```
